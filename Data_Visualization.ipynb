```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/FitnessCustomers.csv"
df1=pandas.read_csv(url)
print(df1)
```

```python
df=pandas.DataFrame(df1)
print(df)
```

```python
seaborn.lineplot(x='Age',y='BMI',data=df)
pyplot.title('Average BMI Rate')
pyplot.show()
```

```python
seaborn.barplot(x='Status',y='Age',data=df)
pyplot.title('Fitness Status')
pyplot.show()
```

```python
seaborn.barplot(x='Age',y='BMI',data=df)
pyplot.title('Average BMI Rate')
pyplot.show()
```

```python
seaborn.scatterplot(x='Age',y='BMI',data=df)
pyplot.title('Average BMI Rate')
pyplot.show()
```

```python
seaborn.histplot(x='Gender',y='Alcohol',data=df)
pyplot.title('Average BMI Rate')
pyplot.show()
```
